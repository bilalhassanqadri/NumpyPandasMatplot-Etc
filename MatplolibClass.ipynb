```python
fig= plt.figure()
# LArge Axes
axes1=fig.add_axes([0,0,1,1]) # 0,0 are the percentage sizes of figur
# and 1,1 is the width * height in percentage
axes1.plot(a,b)
# small axes

axes2=fig.add_axes([0.2,0.2,0.5,0.5]) # so here 0.2 is 20% (x) on graph 
#0.2 is 20% (y) on graph posotion and 0.5 is 50% of width and 0.5 is also
# 50% of width
axes2.plot(x,y);
```

```python
fig= plt.figure()
# LArge Axes
axes1=fig.add_axes([0,0,1,1]) # 0,0 are the percentage sizes of figur
# and 1,1 is the width * height in percentage
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
# small axes
#axes2=fig.add_axes([X,Y,Width,Height]) 
axes2=fig.add_axes([0.1,0.4,0.5,0.5]) # so here 0.2 is 20% (x) on graph 
#0.2 is 20% (y) on graph posotion and 0.5 is 50% of width and 0.5 is also
# 50% of width
axes2.set_xlim(0,10)
axes2.set_ylim(0,15)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.plot(x,y);
```

```python
fig= plt.figure(dpi=200,figsize=(10,3))
axes1= fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

# Subplots multiple plots



```python
a= np.linspace(0,10,11)
b=a**4
```

```python
x=np.arange(0,10)
y=x*2
```

```python
fig,axis=plt.subplots(nrows=2,ncols=2)
axis[0][0].plot(x,y)
axis[0][1].plot(a,b);
plt.tight_layout()
```

# Matplot Styling

```python
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x,x**2,label='X vs X**2')

ax.legend(loc='lower right') # to show the labels Xvs X and X vs X**2
```

```python
fig =plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e6c017',label='X vs X',lw=5,ls='--')#RGB code and color='purple'
# lw parameter is (Line width)
ax.plot(x,x*2,color='blue',label='X vs X**2',linewidth=3,linestyle='-.')
ax.legend();
```
